## glob 이용하기

```python
from glob import glob
```

```python
glob('지역*.xls')
```

```python
stations_files = glob('지역*.xls')
stations_files
```

```python
import pandas as pd

tmp_list = []

for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_list.append(tmp)
    
tmp_list[0]
```

### pandas.concat()
- 동일한 형태의 dataframe을 하나로 합치기

```python
stations = pd.concat(tmp_list)
stations.shape
```

```python
stations.info()
```

info 로 살펴보니 Dtype 이 object! -> 계산을 위해 정수로 바꾸어 주어야함

```python
stations.head()   # 입력안하면 처음부터 5개
```

### loc 이용하기
- iloc
- loc

```python
stations_sel = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '휘발유']]
stations_sel
```

### 칼럼명 변경하기

```python
stations_sel.columns = ['주유소', '주소', '상표', '셀프여부', '가격']
stations_sel
```

### 군 / 구 정보만 추출하기

```python
# '연수구'와 같은 군/구 정보만 추출하기
# '인천 계양구 벌말로 295' ->  '계양구'

gu_list = []

for address in stations['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)

stations_sel['군/구'] = gu_list   # stations_sel에 군/구 컬럼 추가 됨 
stations_sel
```

### 예외처리 하기

```python
# 예외 처리하기

stations_sel.query("가격 == '-'")
```

```python
stations_sel = stations_sel.query("가격 != '-'")  # 가격이 -가 아닌 것만 골라서 가져옴
stations_sel
```

### 가격을 숫자형 정보로 변환해주기
- 변환하지 않으면 그래프를 그릴 때 오류가 생긴다

- SettingWithCopyWarning 나오는 이유 : 일부를 복사하거나 인덱싱 후 값을 수정할 때 발생!!

```python
price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)

stations_sel['가격'] = price_list
stations_sel
```

```python
# SettingWithCopyWarning을 해결하기 위해 .copy() 이용

stations_sel = stations_sel.query("가격 != '-'").copy()
stations_sel
```

```python
# 복사한 걸로 다시 위에랑 같게 해주면 이제 warning 안뜸

price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)

stations_sel['가격'] = price_list
stations_sel
```

```python
stations_sel.info()
```

### reset_index()
- 여러개의 엑셀 파일을 합쳤기 때문에 인덱스가 꼬여 있을 수 있으므로 'reset_index()'를 이용하여 처음부터 다시 기록

```python
stations_sel.reset_index(inplace=True)
stations_sel
```

```python
# 필요 없는 칼럼 지우기

del stations_sel['index']
stations_sel
```

```python
stations_sel_name = stations_sel.set_index('상표')  # 상표를 인덱스로 설정
stations_sel_name
```

##  boxplot 그리기
