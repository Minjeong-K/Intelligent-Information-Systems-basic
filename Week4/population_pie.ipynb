# 파이 차트로 비율 표현하기

## pie() 함수

- 파이 차트 = 원그래프
- 전체 데이터 중 특정 데이터의 비율을 보기 쉽게 표현

```python
import matplotlib.pyplot as plt

plt.pie([100,200])
plt.show()
```

### 레이블 표시하기
- lables !!! (label X)
- 너무 긴 값은 레이블로 전달

```python
plt.rc('font', family = 'Malgun Gothic')
plt.axis('equal')
plt.pie([100,200,300,400], labels=['레이블1','레이블2','레이블3','레이블4'])
plt.show()
```

### 비율 및 범례 표시하기

- autopct='%.1f%%'
- plt.legend()

```python
plt.rc('font', family = 'Malgun Gothic')
plt.axis('equal')
label=['레이블1','레이블2','레이블3','레이블4']
plt.pie([120,250,330,410], labels=label,autopct='%.1f%%') # 소수 첫째자리까지
plt.legend()
plt.show()
```

### 색 정하기
- colors ( 얘도 color 아님! )

```python
plt.rc('font', family = 'Malgun Gothic')
plt.axis('equal')
label=['레이블1','레이블2','레이블3','레이블4']
color = ['darkmagenta', 'deeppink','hotpink','pink']
plt.pie([120,250,330,410], labels=label,autopct='%.1f%%',colors=color)
plt.legend()
plt.show()
```

### 돌출 효과 강조하기
- 강조하기 싶은 데이터의 값 위치에 돌출되는 정도 입력 ( 0 : 돌출 X )
- explode=(0,0,0.1,0)

```python
plt.rc('font', family = 'Malgun Gothic')
plt.axis('equal')
label=['레이블1','레이블2','레이블3','레이블4']
color = ['darkmagenta', 'deeppink','hotpink','pink']
exp = (0,0,0.1,0.3)
plt.pie([120,250,330,410], labels=label,autopct='%.1f%%',colors=color, explode=exp)
plt.legend()
plt.show()
```

# 성별 인구 비율 표현하기

```python
import csv

f = open('incheon_population.csv','r')
data= csv.reader(f)

result = []
m_sum, f_sum = 0, 0
name = '송도1동'

for row in data:
    if name in row[0]:
        for i in range(101):
            m_sum = m_sum + int(row[i+3])
            f_sum += int(row[i+106])

result.append(m_sum)
result.append(f_sum)
print(result)
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.axis('equal')

label=['남','여']
color = ['crimson', 'darkcyan']
plt.pie(result, labels=label,autopct='%.1f%%',colors=color)
plt.legend()
plt.show()
```

### 파이 차트의 시작 각도 정해줌


```python
plt.rc('font', family = 'Malgun Gothic')
plt.axis('equal')

label=['남','여']
color = ['crimson', 'darkcyan']
plt.pie(result, labels=label,autopct='%.1f%%',colors=color, startangle=90)
plt.legend()
plt.show()
```
