# 동계 올림픽 메달 개수 시각화하기
## 201700255 강민정

```python
import pandas as pd
```

```python
url = "https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84"
```

```python
df = pd.read_html(url)
# df
```

```python
# 헤더를 열 이름(columns)으로, 나라 이름을 행 이름(index)로 지정
df = pd.read_html(url, header=0, index_col=0)
# df[0]
```

```python
# 슬라이싱을 활용하여 동계 올림픽에 대한 데이터 추출
a = df[0].iloc[:,5:10]
a
```

```python
# columns 명 변경
a.columns = ['동계 참가 횟수','금','은','동','계']
a
```

```python
# 금메달 기준으로 내림차 순 정렬
a = a.sort_values(by=['동계 참가 횟수','금'],ascending=False)
a
```

### 데이터 시각화하기

```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
```

```python
# 상위 15개국 막대그래프로 표현
a[1:16].plot(kind='bar')
```
