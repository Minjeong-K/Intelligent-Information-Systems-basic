````python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20201024"
html = urlopen(url)

soup = BeautifulSoup(html.read(), "html.parser")

print(soup)
````

````python
title_list = soup.find_all('div',{'class':'tit5'})
print(title_list)
````

````python
for title_div in title_list:
    title = title_div.find('a')
    print(title.text)
````

````python
# div 안에는 text가 하나기 때문에 이렇게 해도 되긴함
for title in title_list:
    print(title.text)
````

````python
title_text_list_1 = []
title_text_list_2 = []
for title in title_list:
    title_text_list_1.append(title.find('a').text)
    title_text_list_2.append(title.text)  # /n 도 같이 들어감
    
print(title_text_list_1)
print(title_text_list_2)
````

### strip()
- **strip()** : 문자열의 왼쪽과 오른쪽에 있는 연속된 모든 공백 제거
- **strip('삭제할 문자들')** : 문자열의 왼쪽과 오른쪽에 있는 연속된 모든 삭제할 문자들 제거

````python
for title in title_text_list_2:
    print(title.strip('\n'))
````

### .string vs .text

텍스트를 추출하는 두 가지 방법

- **.string**
    - 찾고자 하는 태그 ( ```<td>``` ) 하위에 있는 문자열을 객체화
    - 순수 문자열이 아닌 경우, None 반환

- **.text**
    - 해당 태그 안의 모든 텍스트 추출
    - 현재 태그 및 모든 자식(child) 태그를 제거하고 안에 있는 문자열 반환

````python
html_str = """
<html>
<td>some text</td>
<td></td>
<td><p>more text</p></td>
<td>even <p>more text</p></td>   // p태그 안은 순수 문자열 X
</html>
"""

# .string
soup = BeautifulSoup(html_str, "html.parser")
    
tmp = soup.find_all('td')
for t in tmp:
    print(t.string)
````

````python
# .text
soup = BeautifulSoup(html_str, "html.parser")
    
tmp = soup.find_all('td')
for t in tmp:
    print(t.text)
````

````python
# .content
soup = BeautifulSoup(html_str, "html.parser")
    
tmp = soup.find_all('td')
for t in tmp:
    print(t.contents)   # 태그안의 내용을 리스트로 반환
````

### get_text() vs .text

- get_text() 에서 인수 (parameter) 가 없는 경우에는 .text 와 동일

- 가능한 파라미터 : separator, strip, types

````python
# separator
soup = BeautifulSoup(html_str, "html.parser")
    
tmp = soup.find_all('td')
for t in tmp:
    print(t.get_text(separator = '\t'))
````

````python
# strip
for t in tmp:
    print(t.get_text(strip = '\n'))    # t.text.strip('\n') 과 같음
````

## 평점 값 가져오기

````python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20201024"
html = urlopen(url)

soup = BeautifulSoup(html.read(), "html.parser")

point_list = soup.find_all('td',{'class':'point'})
print(point_list)
````

````python
for point in point_list:
    print(float(point.text))  # 문자열 -> 숫자형
````
