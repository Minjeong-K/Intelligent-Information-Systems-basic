- 데이터의 상관관계를 알 수 있음!

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.scatter([1,2,3,4],[10,30,20,40],s=[100,200,250,300]) # s : size의미
plt.show()
```

### 4개의 점 => 위치, 크기가 서로 다른 100개의 점 만들기
- 랜덤 함수 활용하기
    - randint(최소, 최대)

```python
import random

x,y = [],[]
size = []

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x,y,s=size)
plt.show()
```

```python
import random

x,y = [],[]
size = []

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x,y,s=size, c=size, cmap='jet')
plt.colorbar()

plt.show()
```

### 투명도 설정
- alpha : 0 ~ 1

```python
plt.scatter(x,y,s=size, c=size, cmap='jet', alpha=0.5)
plt.colorbar()

plt.show()
```

## 연령대별 성별 비율 산점도 표현

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv','r')
data= csv.reader(f)

male, female = [], []
name = '송도1동'

for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            male.append(int(m))
            female.append(int(f))
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('연령대별 성별 인구 그래프')
plt.style.use('ggplot')
plt.scatter(male, female, c=range(101), alpha=0.5,cmap='jet')
plt.colorbar
plt.xlabel('남성 인구 수')
plt.xlabel('여성 인구 수')
plt.show()
```

### y=cx 형태의 추세선 추가하기

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('연령대별 성별 인구 그래프')
plt.style.use('ggplot')
plt.scatter(male, female, c=range(101), alpha=0.5,cmap='jet')
plt.colorbar

line = range(max(male))
plt.plot(line, line, color='g')

plt.xlabel('남성 인구 수')
plt.xlabel('여성 인구 수')
plt.show()
```

### 원의 크기는 인구수의 제곱근
- 남성과 여성 인구수를 합친 값을 size 리스트에 넣어 점의 크기 표현

```python
import csv
import math
import matplotlib.pyplot as plt

f = open('incheon_population.csv','r')
data= csv.reader(f)

male, female = [], []
size = []
name = '송도1동'

for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            male.append(int(m))
            female.append(int(f))
            size.append(math.sqrt(int(m)+int(f)))
        

plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title('연령대별 성별 인구 그래프', fontsize=15,color='b')
plt.rcParams['axes.unicode_minus'] = False

plt.scatter(male, female, c=range(101),alpha=0.9,cmap='jet')
plt.colorbar()

max_val = max(max(male), max(female))
plt.axis([0,max_val,0,max_val])
line = range(max_val)
plt.plot(line, line, color='g')

plt.xlabel('남성 인구 수')
plt.xlabel('여성 인구 수')

plt.show()      
```
