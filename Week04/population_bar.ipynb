# 막대그래프 그리기

```python
import matplotlib.pyplot as plt

plt.bar([0,1,2,4,6,10],[1,2,3,5,6,7])
plt.show()
```

## bar() 함수
- bar(막대를 표시할 위치, 막대의 높이)
- range() 함수를 이용하여 오름차 순 데이터 입럭

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv','r')
data= csv.reader(f)

result = []
name = '송도1동'

for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            result.append(int(m)+int(f))
            
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 연령별 인구 수')

plt.bar(range(101),result)
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.show()    
```

```python
plt.barh(range(101),result)  # 수평 막대 그래프
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.show()  
```

## 항아리 모양 그래프 그리기

```python
import csv

f = open('incheon_population.csv','r')
data= csv.reader(f)

male = []
female = []
name='송도1동'

for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            male.append(int(m))
            female.append(int(f))
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 연령별 인구 수')

plt.barh(range(101),male)
plt.barh(range(101),female)
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.show()
```

## 하나의 데이터를 음수로 넣어서 방향을 반대로!
# => 항아리 모양!

```python
import csv

f = open('incheon_population.csv','r')
data= csv.reader(f)

male = []
female = []
name='송도1동'

for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            male.append(-int(m))
            female.append(int(f))
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 연령별 인구 수')

plt.rcParams['axes.unicode_minus']=False # 음수 부호 안보일 때 추가
plt.barh(range(101),male, label='남성')
plt.barh(range(101),female, label='여성')
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.legend()
plt.show()
```
