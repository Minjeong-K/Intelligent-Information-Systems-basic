```python
import csv

f = open('incheon_population.csv','r')
data= csv.reader(f)

songdo1_male = []

for row in data:
    if '송도1동' in row[0]:
        for i in row[3:104]:
#             print(i)
            songdo1_male.append(int(i))
print(songdo1_male)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot') # 격자 무늬 스타일 지정
plt.plot(songdo1_male)
plt.show()
```

## 우리 동네 여성 연령별 인구 수 데이터 시각화

```python
import csv

f = open('incheon_population.csv','r')
data= csv.reader(f)

songdo1_female = []

for row in data:
    if '송도1동' in row[0]:
        for i in row[106:207]:
#             print(i)
            songdo1_female.append(int(i))
```

```python
plt.rc('font', family = 'Malgun Gothic') # 폰트 지정
plt.title('송도 1동 여성 연령별 인구 수') # 제목 설정

plt.style.use('ggplot')
plt.plot(songdo1_female)
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.show()
```

## 우리 동네 연령별 인구수(여성+남성) 데이터 시각화하기

```python
import csv

f = open('incheon_population.csv','r')
data= csv.reader(f)

songdo1_male = []
songdo1_female = []
songdo1 = []

for row in data:
    if '송도1동' in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            songdo1_male.append(int(m))
            songdo1_female.append(int(f))
            songdo1.append(int(m)+int(f))
```

### zip 함수
- 동일한 개수로 이루어진 자료형을 묶어주는 역할을 하는 함수

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('송도 1동 연령별 인구 수')

plt.style.use('ggplot')
plt.plot(songdo1)
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.show()
```

- 송도 1동 인구 구조 시각화로 알 수 있는 사실들
    - 50대가 가장 많이 산다
    - 30대는 20대나 50대에 비해 적다
    - 70대 이상은 그리 많지 않다

```python
# 다른 지역에 대해 알아보고 싶으면?

import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv','r')
data= csv.reader(f)

result = []
name = input('인구 구조를 알고 싶은 동 입력 : ')

for row in data:
    if name in row[0]:
        for m,f in zip(row[3:104],row[106:207]):
            result.append(int(m)+int(f))
            
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 연령별 인구 수')

plt.style.use('ggplot')
plt.plot(result)
plt.xlabel('연령')
plt.ylabel('인구 수')
plt.show()    
```
