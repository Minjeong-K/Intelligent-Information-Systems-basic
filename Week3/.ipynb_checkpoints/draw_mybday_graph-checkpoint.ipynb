### 매년 내 생일의 최고 기온 및 최저 기온 데이터 시각화 하기

```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)
next(data)
year = []
high = []
low = []

for row in data:
    if row[-1] != '' :
        mybday = row[0].split('-')
        if int(mybday[0]) >= 1998:                      # 1998년
            if mybday[1] == '05' and mybday[2] == '23': # 5월 23일 생
                year.append(int(mybday[0]))
                high.append(float(row[-1]))
                low.append(float(row[-2]))

f.close()
# print(high)
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic') # 한글 넣기 위해 폰트 설정
plt.title('내 생일의 최고 온도 및 최저 온도 그래프')

plt.plot(year,high,label='최고온도',marker='*', color='r')
plt.plot(year,low, label='최저온도',marker='.',color='b')

plt.legend()

plt.show()
```
