# 그래프에 옵션 추가하기

### 1. 그래프에 제목 넣기

 => plt.title('그래프 제목')


### 2. 그래프에 범례 넣기

 => plot() 함수에 label 이라는 속성의 레이블 값으로 원하는 문자 열 입력

 => plt.legend()
 
     - 범례 위치를 직접 지정하려면?
     
     
     

### 3. 그래프 색상 바꾸기

=> plot() 함수에 color 라는 속성에 원하는 색상 입력

=> 색상을 표현할 때 기본 색 몇가지는 약자로 표기 가능 ( blue 는 color='b' 등... )


### 4. 그래프 선 모양 바꾸기

=> plot() 함수에 linestyle 이라는 속성에 원하는 선 모양 지정

    ( 이때, linestyle 속성 대신 ls라고 작성 할 수도 있음 )

### 5. 마커 모양 바꾸기

=> plot() 함수에 marker 속성을 설정하면 선 형태가 아닌 점 형태의 그래프를 그릴 수 있음


```python
plt.rc('font',family='Malgun Gothic') # 한글 제목을 넣기 위해 폰트 설정
plt.title('플로팅')

# plt.plot([10,20,30,40], color='skyblue', label='범례1') # 색, label 설정
# plt.plot([40,30,20,10], color='pink', label='범례2')

plt.plot([10,20,30,40], color='skyblue', linestyle='--', marker='v', label='범례1') # 색, label 설정, linestyle 설정, marker 설정
plt.plot([40,30,20,10], color='pink', ls=':',marker='o' ,label='범례2')


plt.legend()
# plt.legend(loc=9)   # 범례 위치 설정

plt.show()
```

# 기온 변화 그래프 그리기

## 리스트에 데이터 저장하기

```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)

header = next(data)
result = []

for row in data:
#     print(row)
    if row[-1] != '' :
        result.append(float(row[-1]))

f.close()

# print(result)
print(len(result))
```

## 데이터 시각화 하기

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus']=False  # 마이너스 부호가 표현되지 않는 경우
plt.plot(result, color='red')
plt.show()
```

## 날짜 데이터 추출하기

### 문자열을 분리하는 split() 함수

```python
date = '1997-10-01'
print(date.split('-'))
```

```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)
next(data)
result = []

for row in data:
#     print(row)
    if row[-1] != '' :
        date0913 = row[0].split('-')
        if date0913[1] == '09' and date0913[2] == '13': # 9월 13일에 해당한다면
            result.append(float(row[-1]))

f.close()

print(result)
```

```python
plt.plot(result, color='hotpink')
plt.show()
```
