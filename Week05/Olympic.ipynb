```python
# 헤더를 열 이름(columns)으로, 나라 이름을 행 이름(index)로 지정
df = pd.read_html(url, header=0, index_col=0)
df[0]
```

```python
# 슬라이싱을 활용하여 하계 올림픽에 대한 데이터 추출
a = df[0].iloc[:,:5]
a
```

```python
a.columns = ['하계 참가 횟수','금','은','동','계']
a
```

```python
# 금메달 기준으로 내림차 순 정렬
a = a.sort_values(by=['금'],ascending=False)
a
```

```python
a.to_csv('summer_medal.csv') # csv 파일로 저장
```
